```python
# write your code here, you can add more cells if needed
import pandas as pd
df = pd.read_csv("pre-course_survey.csv")
df.head()
```

## 2. Preprocess the data (from week3 exercise)

1. Rename all columns
2. Drop nan
3. Drop duplicates

```python
def process_data(dataframe):
    """
    This function performs the necessary preprocessing steps on a given dataframe.
    The input here is a dataframe, not a filename. If you want to use a filename as input, 
    you need to add a line of code to read the CSV file into a DataFrame.

    Parameters:
        dataframe: a pandas DataFrame object.

    Returns:
        dataframe: the processed pandas DataFrame object.
    """

    # Rename the columns
    old_cols = dataframe.columns
    new_cols = ["timestamp", "python", "datascience", "git", "jupyter", 
                "reproducibility", "nlp", "socialmedia_data", "goals", 
                "skills", "concerns"]
    cols_dict = dict(zip(old_cols, new_cols))
    dataframe.rename(columns=cols_dict, inplace=True)

    # Remove rows in the DataFrame that contain missing values.
    dataframe.dropna(inplace=True)

    # Remove duplicate rows from the DataFrame.
    dataframe.drop_duplicates(inplace=True)

    # Recode 'Yes' and 'No' values in the DataFrame to 1 and 0 respectively.
    dataframe.replace({"Yes":1, "No":0}, inplace=True)
    
    # Return the processed DataFrame.
    return dataframe
```

```python
df_new = process_data(dataframe=df)
df_new.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Line Plot

```python
# Create a line plot using seaborn's lineplot function. 
# The x-axis represents the 'python' column from 'df_new', and the y-axis represents the 'reproducibility' column.
sns.lineplot(x="python", y="reproducibility", data=df_new) 
# Set the title of the plot to "Line Plot".
plt.title("Line Plot")
# Change the label of the x-axis to "Python Knowledge".
plt.xlabel("Python Knowledge")
# Change the label of the y-axis to "Reproducibility Knowledge".
plt.ylabel("Reproducibility Knowledge")
# Display the plot. Without this, the plot object is created but not displayed.
plt.show()
```

### Scatter Plot 

```python
# write your code here, you can add more cells if needed
sns.scatterplot(x="datascience", y="reproducibility", hue="git", data=df_new)
plt.title("Scatter Plot")
plt.xlabel("Data Science Knowledge")
plt.ylabel("Reproducibility Knowledge")
plt.legend(title="Git Knowledge")
plt.show()
```

###  Violin Plot (You can choose other plots)

```python
# write your code here, you can add more cells if needed
sns.violinplot(x='socialmedia_data', y='datascience', data=df_new)
plt.title("Scatter Plot")
plt.xlabel("Social Media Data Knowledge")
plt.ylabel("Data Science Knowledge")
plt.show()
```
